#T20 World Cup ***2022 ***Analysis

```python
import pandas as pd
import json
```

## match results (file processing)

```python
with open('/content/t20_wc_match_results.json') as f:
    data=json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match id'},axis=1 ,inplace= True)
df_match.head()
```

**Batting summary** (file processing)

```python
df_match.to_csv('fact_match_summary.csv',index=False)
```

```python
{
    "Nimbia VS Sri Lanka": "T20I # 1823",
    "Sri Lanka VS Nimbia": "T20I # 1823",
    "Sri Lanka VS England": "T20I # 1824",
    "England VS Sri Lanka": "T20I # 1824",
    "Sri Lanka VS Pakistan": "T20I # 1825",
    "Pakistan VS Sri Lanka": "T20I # 1825",
    "Sri Lanka VS West Indies": "T20I # 1826",
    "West Indies VS Sri Lanka": "T20I # 1826",
    "Sri Lanka VS Netherlands": "T20I # 1827",
    "Netherlands VS Sri Lanka": "T20I # 1827",
}
```

```python
match_ids_dict={}

for index,rows in df_match.iterrows():
    key1=rows['team1'] + ' Vs ' + rows['team2']
    key2=rows['team2'] + ' Vs ' + rows['team1']
    match_ids_dict[key1]=rows['match id']
    match_ids_dict[key2]=rows['match id']
match_ids_dict
```

```python
with open('/content/t20_wc_batting_summary.json') as f:
    data=json.load(f)

    all_records=[]
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting=pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not out")
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head()
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('',''))
df_batting.head(10)
```

```python
df_batting['match id']=df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('fact_bating_summary.csv',index=False)
```

##Bowling Summary(file processing)

```python
with open('/content/t20_wc_bowling_summary.json') as f:
    data=json.load(f)
all_records=[]
for rec in data:
        all_records.extend(rec['bowlingSummary'])

df_bowling=pd.DataFrame(all_records)
df_bowling.head()
```

```python
df_bowling['match id']=df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('fact_bowling_summary.csv',index=False)
```
